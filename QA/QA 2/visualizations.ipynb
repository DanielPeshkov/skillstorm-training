```python
import matplotlib.pyplot as plt
```

```python
y = [109, 244]
x = ['Agg then Join', 'Join then Agg']

fig, ax = plt.subplots(figsize=(8, 8))
plt.bar(x, y, color='skyblue', ec='k', lw=1)
ax.set_ylabel('Time in MS', fontsize=16)
ax.set_title('Aggregating Before vs After Joining', fontsize=20)

plt.savefig("agg_bar.png", bbox_inches='tight')

plt.show()
```

```python
x = [15, 30, 60]
y1 = [4, 5, 11]
y2 = [12, 15, 33]

fig, ax = plt.subplots(figsize=(6, 6))
plt.plot(x, y1, label="Indexed", lw=4, c='green')
plt.plot(x, y2, label="Not Indexed", lw=4, c='firebrick')
plt.legend()
ax.set_ylabel('Query Time in MS', fontsize=16)
ax.set_xlabel('Thousand Rows In Table', fontsize=16)
ax.set_title('Indexed vs Not Indexed Query Time', fontsize=20)

plt.savefig("index_line.png", bbox_inches='tight')

plt.show()
```

```python
y = [12, 88]
x = ['Not Exists', 'Not In']

fig, ax = plt.subplots(figsize=(8, 8))
plt.bar(x, y, color='skyblue', ec='k', lw=1)
ax.set_ylabel('Time in MS', fontsize=16)
ax.set_title('Not In vs Not Exists Execution Time', fontsize=20)

plt.savefig("in_exists_bar.png", bbox_inches='tight')

plt.show()
```

```python
y = [28, 52]
x = ['Exists', 'In']

fig, ax = plt.subplots(figsize=(8, 8))
plt.bar(x, y, color='skyblue', ec='k', lw=1)
ax.set_ylabel('Time in MS', fontsize=16)
ax.set_title('In vs Exists Execution Time', fontsize=20)

plt.savefig("not_in_exists_bar.png", bbox_inches='tight')

plt.show()
```

```python
x = ['Explicit Transaction', 'Implicit Transaction']
l = [430, 2600]
h = [100, 600]

fig, ax = plt.subplots(figsize=(8, 8))
plt.bar(x, l, color='skyblue', ec='k')
plt.bar(x, h, bottom=l, ec='k')
ax.set_ylabel('Time in MS', fontsize=16)
ax.set_title('Insertion of 50k Rows', fontsize=20)

plt.savefig("transaction_bar.png", bbox_inches='tight')

plt.show()
```
